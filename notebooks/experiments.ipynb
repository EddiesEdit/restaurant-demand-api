Restaurant Food Sales - Synthetic Time Series Data


This dataset contains daily sales records of food dishes across multiple restaurant localities and is designed for time series demand forecasting. Each row represents the number of units sold (units_sold) for a specific dish on a particular date, along with rich contextual features that describe temporal patterns, restaurant attributes, dish characteristics, pricing signals, and external influences. Time-based attributes such as weekday, month, season, is_weekend, and working_day help capture weekly and seasonal sales trends. Dish-related attributes like category, veg_nonveg, ingredient, foodtype, meal_type, spice_level, and size represent culinary characteristics that influence customer preference. Restaurant-level variability is modeled through restaurant_locality, while features such as price_per_unit, price_mean, rating, and dish_popularity contribute economic and quality-related factors affecting demand. Additionally, features like is_holiday, holiday_religion, and is_festival represent demand spikes during cultural and religious events. Together, the dataset provides a realistic multi-dimensional view of restaurant sales behavior over time, making it well-suited for building robust time series models like LSTM that can learn both temporal dependencies and complex feature interactions.
